## 4. How to Interpret the Intercept in a Regression Model 

Suppose wed like to fit a simple linear regression model using hours studied as a predictor variable and exam score as the response variable.
We collect this data for 50 students in a certain college course and fit the following regression model:

Exam score = 65.4 + 2.67(hours)

The value for the intercept term in this model is 65.4. This means the average exam score is 65.4 when the number of hours studied is equal to zero.This makes sense to interpret since its plausible for a student to study for zero hours in preparation for an exam.



Source: https://www.statology.org/intercept-in-regression/


## 5. What is RSquare?

- R Square is prportion of variance of the dependent varible explained by Idepednet variable
- If R Suare is 80%: It means 80% of the value of the Y is because of present independ varibale on X axis
- With adding more independent variables, R Square can only go up, never down



## 6. What is Adjusted R Square

- One limitation of R Square is that it always incraesed if a independent variable is added in the model. Sometime if a useless variable ia added. It still increases R Square. Hence it overestimates the prediction.

- Adjusted R2 attempts to correct for this overestimation. Adjusted R2 might decrease if a specific effect does not improve the model.(in our case if added column is Useless meaning not related to target)

- Define decision trees including classification and regression ( for umerical Prediction) trees.
- Explain main ideas of decision trees based on recursive partitioning and tree pruning (stop groeth of Tree)
- When finally a tree is grown using recursive partitioning. This tree is used for classification in new record. Hence this grown tree is called Claasification Tree because it hepls to classify the target value.


- Please read third point again and dont get confused. A tree is always made using recursive partitioning ( making classes at each step) . Once it is made, It is used to classfiy target value in new records, hence called classsfication Tree or Decision Tree.



## 14. What are the priciples ( Key Ideas) of Decision Tree?

- Recursive partitioning:  

Repeatedly split the records into two subsets so as to achieve maximum homogeneity within the each of the new subsets

- Stopping tree growth:

A fully grown tree is too complex and will overfit. Simplify the tree by pruning peripheral branches to avoid overfitting


## 15. What is Recursive partitioning, and how it is achieved conceptually ?

Recursive partitioning is the step-by-step process by which a decision tree is constructed by either splitting or not splitting each node on the tree into two daughter nodes w.r.t to a prediction at each branch (depth).

## 20. What is Cross Validation in Decision Tree?

- A different partition into training/validation from the roiginal dataset could lead to a different initial split. This can produce a very different tree from the first training/validation partition.

- So if mutiple partitions production different accuracies. It is important to check variability in the accuracies. If there is major change it means our model is giving different results hence we need to switch to some othe rmethod. If variability is less, It means our model is performing constant for all kind of partitions.

- Hence in CV, we make differents paretitions and check validate our model in terms of variability ( Standar deviation of accuracies)
