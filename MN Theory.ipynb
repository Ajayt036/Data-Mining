## 4. How to Interpret the Intercept in a Regression Model 

Suppose we’d like to fit a simple linear regression model using hours studied as a predictor variable and exam score as the response variable.
We collect this data for 50 students in a certain college course and fit the following regression model:

Exam score = 65.4 + 2.67(hours)

The value for the intercept term in this model is 65.4. This means the average exam score is 65.4 when the number of hours studied is equal to zero.This makes sense to interpret since it’s plausible for a student to study for zero hours in preparation for an exam.

## 47. What is the Concept/Principle of Neural Network?

It is an intertive Concept:

- Combine the input information in a flexible neural net model that captures complex relationships between input variables and the outcome (response).
- Model coefficients/parameters are continually tweaked in an iterative process
- Network’s interim performance in classification and prediction informs successive tweaks
- In neural networks, the user is not required to specify the correct form of relationships; the network tries to learn about such relationships   

## 48. What is Neural Network Structure?

Neural network structure is typically a feedforward network which consists of  multiple layers
- Input layer of raw observations consisting of nodes  or neurons 
- Output layer – the last layer(s) in the neural networks consisting of one or several nodes
- Hidden layer(s) – layer(s) between input and output that accept the input values and successive layers (nodes) receive inputs from previous layers 

Feedforward network is a fully connected network with a one-way flow and no cycles.

## 2. How many types of Clusterning are there?

1. Hierarchical methods ( Follows an order to make cluster )

- Agglomerative method begins with n clusters (for n observations) and sequentially merge similar clusters until a single cluster is obtained
- Divisive method works in the opposite direction, starting with one cluster that includes all observations
- Both are specifically useful when the goal is to arrange the clusters into a natural hierarchy

2. Nonhierarchical methods ( Just make cluster based on specified number )

- k-Means Clustering (a most popular nonhierarchical method) are used for a prespecified number of clusters with observations assigned to each cluster
- These methods are generally less computationally intensive and are, therefore, preferred with very large datasets

## 4. How do we use Meaure distance in Forming group of clusters in Agglomerative clustering?

1. First We form clusters: 

- Just based on the Eculidean distance from measuremnts, we form cluster.
- The close value of distance makes clusters.


2. Distance between clusters: Once the clusters are formed, we see how far they are from each other. can we merge them if they are close, we check that as well. There are many ways to check this closness(measure distance) between clusters.


A. Minimum distance is a good choice for clustering chain- or sausage-like clusters 
- Clustering crops planted in long rows    
-E.g., clustering disease outbreak along waterways (e.g., rivers)

B. Maximum distance and average distance are good choices for clustering if clusters are more likely to be spheric
- E.g., customers clustered on the basis of numerous attributes
- Good default choices since most clusters tend to be spherical
